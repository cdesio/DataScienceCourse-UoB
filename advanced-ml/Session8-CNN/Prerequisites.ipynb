## 1. Linear Algebra Essentials

**Why it matters:** Neural networks use matrices and vectors to represent data and parameters. Each layer performs matrix multiplications, so understanding these operations is critical.

### Vectors and Matrices
A **vector** is an ordered list of numbers, written vertically or horizontally.  
A **matrix** is a grid of numbers arranged in rows and columns.

Example:

$$
\mathbf{x} = 
\begin{bmatrix}
x_1 \\
x_2 \\
x_3
\end{bmatrix}, \quad
\mathbf{W} = 
\begin{bmatrix}
w_{11} & w_{12} \\
w_{21} & w_{22} \\
w_{31} & w_{32}
\end{bmatrix}
$$

Vectors and matrices are the “containers” that hold data and weights in your model.

## Matrix–Vector Product
This operation combines input features and weights:

$$
\mathbf{z} = \mathbf{W}^T \mathbf{x} + \mathbf{b}
$$

where $\mathbf{W}$ represents weight connections, $\mathbf{x}$ is the input, and $\mathbf{b}$ is the bias.



### Matrix–Vector Multiplication Explained

Matrix–vector multiplication is a key operation in machine learning and neural networks because it describes how inputs combine with weights to produce outputs in each layer.

***

### a. What It Is
If you have a matrix $A$ of size $m \times n$ and a column vector $x$ with $n$ elements, you can multiply them only when the number of **columns in $A$ equals** the number of **rows in $x$**.

The result $b = A x$ is a **new vector** of length $m$.

Formally:

$$
A = \begin{bmatrix}
a_{11} & a_{12} & \ldots & a_{1n} \\
a_{21} & a_{22} & \ldots & a_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
a_{m1} & a_{m2} & \ldots & a_{mn}
\end{bmatrix}, \quad
x = \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix}
$$

Then:

$$
A x = \begin{bmatrix}
a_{11}x_1 + a_{12}x_2 + \ldots + a_{1n}x_n \\
a_{21}x_1 + a_{22}x_2 + \ldots + a_{2n}x_n \\
\vdots \\
a_{m1}x_1 + a_{m2}x_2 + \ldots + a_{mn}x_n
\end{bmatrix}
$$

Each element of the result is the **dot product** between one row of $A$ and the vector $x$.

*(Source adapted from [Math Insight](https://mathinsight.org/matrix_vector_multiplication).)*[1]

***

### b. Example
Let:

$$
A = \begin{bmatrix} 1 & -1 & 2 \\ 0 & -3 & 1 \end{bmatrix}, \quad x = \begin{bmatrix} 2 \\ 1 \\ 0 \end{bmatrix}
$$

Then multiplying gives:

$$
A x = \begin{bmatrix} 1(2) + (-1)(1) + 2(0) \\ 0(2) + (-3)(1) + 1(0) \end{bmatrix} = \begin{bmatrix} 1 \\ -3 \end{bmatrix}
$$

So the result is a 2-element vector.

***

### c. Why It Matters
In neural networks:
- $A$ represents the **weight matrix**.
- $x$ is the **input vector**.
- The output $A x$ gives the weighted sum passed through an activation function.

This operation is how each layer of a network transforms its inputs into new representations.

## Linear Transformations
Each layer in a neural network is a *linear transformation* followed by a *non-linear activation* — visually, this maps points from one space to another.

### Helpful Concepts
- **Transpose:** $A^T$ flips rows and columns.  
- **Dot Product:** Combines vectors to produce a single number, $x \cdot y = \sum_i x_i y_i$.  
- **Identity Matrix:** Acts like “1” in multiplication, $I x = x$.

## 2. Gradient Descent Intuition

**Why it matters:** Training means finding parameters (weights) that minimize error. Gradient descent is the method used to do this.

### The Idea
Imagine a bowl-shaped loss surface. The goal is to move downhill to the lowest point (minimum loss).

We update weights as:

$$
w_{ij} := w_{ij} - \eta \frac{\partial L}{\partial w_{ij}}
$$

where $\eta$ is the *learning rate*, controlling how large each adjustment step is.

### Chain Rule and Backpropagation
Backpropagation applies the chain rule from calculus to efficiently calculate gradients for all weights:

$$
\frac{\partial L}{\partial w_{ij}} = 
\frac{\partial L}{\partial a_j} \cdot 
\frac{\partial a_j}{\partial z_j} \cdot 
\frac{\partial z_j}{\partial w_{ij}}
$$

## 3. Machine Learning Background

### Supervised Learning
Uses labeled data: pairs of inputs and desired outputs.  
Goal: learn a function $f(x) \approx y$.

### Loss Function
Measures how wrong the predictions are.  
For example, binary classification often uses **cross-entropy loss**:

$$
L(y, \hat{y}) = -y \log(\hat{y}) - (1 - y)\log(1 - \hat{y})
$$

The model updates its parameters to minimize this loss.

### From Linear Models to Neural Networks
- The **Perceptron** is a single neuron, similar to logistic regression.  
- A **Multi-Layer Perceptron (MLP)** stacks several layers, each adding non-linear transformations to capture complex relationships.
