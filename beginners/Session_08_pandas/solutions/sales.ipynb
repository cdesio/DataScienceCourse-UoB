```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Generate random sales data
num_products = 10000
product_ids = np.arange(1, num_products + 1)
categories = np.random.choice([1, 2, 3, 4], size=num_products)
# 1: Electronics, 2: Clothing, 3: Home, 4:  Sports
prices = np.zeros(num_products)

# Assign different price ranges for each category
electronics_price_range = (50, 500)
clothing_price_range = (20, 200)
home_price_range = (100, 1000)
sports_price_range = (10, 300)

for i in range(num_products):
    category = categories[i]
    if category == 1:
        prices[i] = np.random.uniform(
            electronics_price_range[0], electronics_price_range[1])
    elif category == 2:
        prices[i] = np.random.uniform(
            clothing_price_range[0], clothing_price_range[1])
    elif category == 3:
        prices[i] = np.random.uniform(home_price_range[0], home_price_range[1])
    elif category == 4:
        prices[i] = np.random.uniform(
            sports_price_range[0], sports_price_range[1])

# Generate sales quantities and calculate revenue
sales_quantities = np.random.randint(1, 20, size=num_products)
revenues = prices * sales_quantities

# Create a DataFrame to store the sales data
data = pd.DataFrame({
    'ProductID': product_ids,
    'Category': categories,
    'Price': prices,
    'Quantity': sales_quantities,
    'Revenue': revenues
})

# Save the sales data to a CSV file
data.to_csv('sales_data.csv', index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Task 1: Load the dataset
data = np.genfromtxt('sales_data.csv', delimiter=',', skip_header=True)

```

```python
#columns: 0: productID, 1: category, 2: price, 3: quantity, 4: revenue
# Task 2: Calculate statistics
total_transactions = data.shape[0]
total_products_sold = np.sum(data[:, 3])
total_revenue = np.sum(data[:, 4])

average_price = np.mean(data[:, 2])
most_expensive_product = data[np.argmax(data[:, 2]), 0]

# Task 3: Find top-selling product category
category_sales = np.unique(data[:, 1], return_counts=True)
category_names = {1: "Home", 2: "Beauty", 3: "Electronics", 4: "Clothing"}
top_category = category_names[category_sales[0][np.argmax(category_sales[1])]]
# Task 4: Filter the dataset
filtered_data = data[data[:, 3] >= 10]
np.savetxt('filtered_data.csv', filtered_data, delimiter=',', fmt='%.2f')

# Task 5: Calculate total revenue for filtered dataset
filtered_revenue = np.sum(filtered_data[:, 4])

# Task 6: Calculate average price for filtered dataset
filtered_average_price = np.mean(filtered_data[:, 2])



# Summary and Insights
print("Summary:")
print("------------------------------")
print("Total Products Sold:", total_products_sold)
print("Total Revenue Generated: $", total_revenue)
print("Average Price per Product: $", average_price)
print(f"Most Expensive Product: no. {int(most_expensive_product)}")
print("Top-Selling Category:", top_category)
print("Filtered Dataset: 'filtered_data.csv'")
print("Filtered Revenue: $", filtered_revenue)
print("Filtered Average Price: $", filtered_average_price)

print("\nInsights:")
print("------------------------------")
print("1. The company sold a total of", total_products_sold,
      "products, generating $", total_revenue, "in revenue.")
print(f"2. The average price per product is ${average_price}, with the most expensive product being {most_expensive_product}.")
print("3. The top-selling product category is",
      top_category, "with the highest sales quantity.")
print("4. The filtered dataset includes products with a quantity of 10 or more.")
print("5. The total revenue for the filtered dataset is $", filtered_revenue,
      "with an average price per product of $", filtered_average_price, ".")
print("6. The revenue generated by each product category can be visualized using the scatter plot.")
```

```python
# Task 7: Visualize revenue by product category

revenues = data[:, 4]
plt.figure(figsize=(10, 6))
colors = ['red', 'blue', 'green', 'gold']
categories = data[:,1]
for i, cat in enumerate(categories):
      plt.plot(i, revenues[i], 
               linestyle='-.', marker='s', linewidth=.2, c=colors[int(cat-1)], label=cat)

plt.xlabel('Product no.')
plt.ylabel('Revenue')
plt.title('Revenue Generated by Each Product')
plt.tight_layout()
# plt.legend()
plt.show()
```

```python
categories = np.unique(data[:, 1]) 
color = ['#990F02', '#241571', '#FF4F00', '#028A0F']
for x , category in enumerate(categories):
    data = data[data[:, 1] == category]
    revenue = data[:,4]
    plt.scatter(range(len(revenue)), revenue,color=color[x], label=category, alpha=.5)
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.legend(title = 'Category')
plt.show()
```
