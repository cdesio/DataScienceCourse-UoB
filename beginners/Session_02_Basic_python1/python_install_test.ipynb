### ✅ **Python Installation Test Notebook with Conda Install Suggestions**

```python
# Test 1: Basic Python functionality
print("Hello, world!")
```

```python
# Test 2: Check Python version
import sys
print("Python version:")
print(sys.version)
```

```python
# Test 3: Check if NumPy is installed
try:
    import numpy as np
    print("NumPy is installed. Version:", np.__version__)
    print("NumPy array example:", np.array([1, 2, 3]) + 1)
except ImportError:
    print("NumPy is not installed.")
    print("To install, run: !conda install numpy -y")
```

```python
# Test 4: Check if Pandas is installed
try:
    import pandas as pd
    print("Pandas is installed. Version:", pd.__version__)
    print("Pandas DataFrame example:\n", pd.DataFrame({'A': [1, 2], 'B': [3, 4]}))
except ImportError:
    print("Pandas is not installed.")
    print("To install, run: !conda install pandas -y")
```

```python
# Test 5: Check if Matplotlib is installed
try:
    import matplotlib
    import matplotlib.pyplot as plt
    print("Matplotlib is installed. Version:", matplotlib.__version__)
    plt.plot([0, 1], [0, 1])
    plt.title("Test Plot")
    plt.show()
except ImportError:
    print("Matplotlib is not installed.")
    print("To install, run: !conda install matplotlib -y")
```
